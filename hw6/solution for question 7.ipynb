```python
#第7题  用ax+b线性来拟合y=x^2
import numpy as np 
import matplotlib.pyplot as plt

data = np.random.rand(2,1000)       #(用两个点来模拟的模型，详情见参考答案)
w = 0
b = 0
for i in range(1000):
    w0 = data[0][i]+data[1][i]
    b0 = -data[0][i]*data[1][i]
    w += w0 
    b += b0 
print(w/1000, b/1000)
```

```python
#第7题图像验证：
from sklearn.linear_model import LinearRegression  #引入线性回归模型

data=np.random.rand(10000,1)
y=data**2

fig,ax=plt.subplots(figsize=(12,8))
for i in range(10000):
    ax.scatter(data[i][0],data[i][0]**2,s=50,marker='x',c='r')
model=LinearRegression()


model.fit(data,y)
w=model.coef_.reshape(1,1)
b=model.intercept_
print(w)
print(model.intercept_)

x=np.arange(0,1,0.01)
y=w[0][0]*x+b
ax.plot(x,y,c='b')

```
